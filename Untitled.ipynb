```python
data.diagnosis=[ 1 if each =="M" else 0 for each in data.diagnosis]
```

```python
y=data.diagnosis.values
```

```python
x_data=data.iloc[ :,1:3].values
```

```python
x=(x_data-np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
y_head=knn.predict(x_test)
y_head
```

```python
print("when k is {}, accuracy score of knn classification {}".format(3,knn.score(x_test,y_test)))
```

```python
test_accuracy=[]
for each in range(1,15):
    knn2=KNeighborsClassifier(n_neighbors=each)
    knn2.fit(x_train,y_train)
    test_accuracy.append(knn2.score(x_test,y_test))
```

```python
plt.figure(figsize=(5,5))
plt.plot(range(1,15),test_accuracy)
plt.title("K values VS accuracy")
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.grid()
plt.show()
print("best accuracy score is {} with K={}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
```
